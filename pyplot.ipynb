```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('data/Wine_data.xlsx')  
df
```

```python
# histogram of 'quality' scores
data=df['quality']
fig, axs = plt.subplots(1, 1, figsize =(10, 7),tight_layout = True)
axs.grid(b = True, color ='grey', linestyle ='-.', linewidth = 0.5, alpha = 0.6)
plt.xlabel("Quality Score")
plt.ylabel("Frequency")
plt.hist(data, color="r", edgecolor ="black")
```

```python
# comaparing alcohol with quality score
# https://www.geeksforgeeks.org/matplotlib-pyplot-scatter-in-python/
x = df['quality']
y = df['alcohol']
# bit of formatting - explain?
plt.scatter(x,y, c="r", linewidths = 1, marker ="^", 
            edgecolor ="green", s = 30)
```

```python
# Ex. 1

# Create a figure and an axes with pyplot.subplots()
fig,axs = plt.subplots(figsize=(15,7)) 
# creates 3 variables with random data - 100 randomly distibuted points with a centre of 0 and SD of 1
data1 = np.random.normal(0,1,100)
data2 = np.random.normal(0,1,100)
data3 = np.random.normal(0,1,100)
# fixes the range on x and y axes
x_ax = np.arange(0,100,10)
y_ax = np.arange(-3,3,1)
# plots the data of each variable
# specifies the 'marker' for each
axs.plot(data1,marker="o", label='data1')
axs.plot(data2,marker="*", label='data2')
axs.plot(data3,marker="^", label='data3')
# defines x and y ticks and labels
# rotation???
axs.set_xticks(x_ax)
axs.set_xticklabels(labels=x_ax,rotation=45)
axs.set_yticks(y_ax)
axs.set_yticklabels(labels=y_ax,rotation=45)
axs.set_xlabel("X label")
axs.set_ylabel("Y label")
# creates an overall table
axs.set_title("Title")
# creates a background grid
axs.grid("on")
plt.legend()
# creates the plot
plt.show()
```

#### Ex. 2

```python
# Create the data
x = np.cos(np.linspace(0, 2, 100))
# Create a figure and an axes with pyplot.subplots()
fig, ax = plt.subplots()  
# Plot some data on the axes
ax.plot(x, x, label='linear')  
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')
# add x and y lables
ax.set_xlabel('x label')
ax.set_ylabel('y label')
# Add a title to the axes
ax.set_title("Simple Plot")  
# Add a legend
ax.legend();  
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.2)

plt.figure()

plt.subplot(2,2,1)
plt.plot(t1, f(t1), 'black')

plt.subplot(2,2,2)
plt.plot(t2, np.tan(2*np.pi*t2), 'r--')

plt.subplot(2,2,3)
plt.plot(t2, np.exp(t2), 'g^')

plt.subplot(2,2,4)
plt.plot(t2, np.cos(2*np.pi*t2), 'orange');
plt.show()
```

#### Ex. 9

```python
# adapted from: https://matplotlib.org/stable/tutorials/introductory/pyplot.html#sphx-glr-tutorials-introductory-pyplot-py

# make up some data in the open interval (0, 1)
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure()

# linear
plt.subplot(221)
plt.plot(x, y, color="red")
plt.yscale('linear')
plt.title('linear',color="red")
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y, color="green")
plt.yscale('log')
plt.title('log',color="green")
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean(),color="blue")
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog',color="blue")
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y, color="orange")
plt.yscale('logit')
plt.title('logit',color="orange")
plt.grid(True)
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)
plt.tight_layout()
plt.show()
```
