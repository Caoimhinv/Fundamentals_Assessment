#### What is Matplotlib?
**Matplotlib** is a comprehensive library for creating visualizations and plots in the Python programming language. It allows for static, animated, and interactive plots to be created with minimal lines of code. It was developed as an open-source alternative to **MATLAB***. It is dependent on the scientific computing package package **NumPy**.

#### What is Pyplot?
**Pyplot** is a module within matplotlib that provides a MATLAB-style interface. It contains a multitude of functions for the creation and manipulation of figures and plots, and all the elements contained within.

*MATLAB is a proprietary multi-paradigm programming language and numeric computing environment developed by MathWorks. MATLAB allows matrix manipulations, plotting of functions and data, implementation of algorithms, creation of user interfaces, and interfacing with programs written in other languages. https://en.wikipedia.org/wiki/MATLAB

For this assignment we are concentrating on matplotlib.pyplot. I will begin with an overview of the syntax and structure of how the module works, and then we will look at a few plots in detail. We begin by importing all of the necessary libraries. 

```python
# numerical arrays
import numpy as np

# plotting
import matplotlib.pyplot as plt

# dataframes
import pandas as pd
```

The most basic and best place to start is with .plot(). Here we create 2 arrays - one for each axis - and then use the plot() function to visualise them.

```python
# x axis is an array of 1 and 10
x = np.array([1,10])

# y axis is an array of 1 and 100
y = np.array([2,25])

# creates the plot
plt.plot(x, y)
```

The function plots the points at 1,2 and then at 10,25 and draws a line between them. If we want to just have the popints we insert an 'o' into the plot function as below.

```python
# creates the plot
plt.plot(x, y, 'o')
```

The 'o' could equally be any one of a multitude of **markers**. Below I've created another plot but this time showing a few other possible markers, and also demonstrating that we can plot multiple sets of arrays on the same figure. The full list of markers is here - https://matplotlib.org/stable/api/markers_api.html. 
We can also specify the markersize e.g. 'ms=12' (or markersize=12). The different arrays are given default colours but this can also be specified. We do this by putting the color 'code' in front of the marker symbol, e.g. 'cD' will produce a cyan diamond. We can change the marker face colour but keep the edgecolour by using e.g. mfc='y' (or markerfacecolor='y'). The marker edge color and edge width are also manipulable with 'mec' (or markeredgecolor) and 'mew' (or markeredgewidth) respectively. The base colors are:
- b = blue
- r = red
- g = green
- c = cyan
- m = magenta
- y = yellow
- k = black
- w = white  

On top of this there are a huge amount of other named colors. Full list here - https://matplotlib.org/stable/gallery/color/named_colors.html

```python
# creates arrays
a = np.array([5,11])
b = np.array([7,19])
c = np.array([4,14])
d = np.array([21,23])

# creates the plot
# mfc='y' returns a marker face color of yellow
plt.plot(a, b, '*', mfc='y', ms=25)

# specifies color for marker edge color and width
plt.plot(x, y, 'P', mec='darkslategrey', mew=2, ms=16)

# 'cD' returns a cyan diamond marker
plt.plot(c, d, 'cD', ms=12)
```

A lot of the same parameters are possible with the linestyle. We can demonstrate this using the same plot but removing the marker parameters. Here we specify the linestyle (ls), linewidth (lw), and color.

```python
# creates the plot
plt.plot(a, b, ls=':', lw=10, color='slategrey')
plt.plot(x, y, ls='--', lw=5, color='hotpink')
plt.plot(c, d, ls='-.', lw=2, color='aqua')
```

To further customise our plot we can add x and y labels, a title, a legend, a background grid, and increase the size of the plot. The following brings it all together!

```python
# sets the overall plot size
plt.figure(figsize=(15, 10))

# sets the x and y labels and specifies their sizes
plt.xlabel("something", size=20)
plt.ylabel("something else", size=8)

# sets the title and specifies size
plt.title("some somethings", size=30)

# creates a background grid and specifies the colour
plt.grid(True, color='lightsalmon')

# plots the various markers and line and specifies individual style elements
plt.plot(a, b, ls=':', lw=15, color='slategrey', label='A&B')
plt.plot(x, y, ls='--', lw=5, color='hotpink', label='X&Y')
plt.plot(c, d, ls='-.', lw=2, color='aqua', label='C&D')
plt.plot(4,20, 'ro', ms=22, label='4&20')
plt.plot(12,5, '*', mfc='y', ms=25, label='12&5')
plt.plot(2,25, 'P', mec='darkslategrey', mew=2, ms=25, label='2&25')

# creates a legend and specifies the size
plt.legend(fontsize=18)
```

This is just an overview of some of the possibilites and obviously not the prettiest plot in the world! I will introducing more as we go through the other types of plots available.

### Scatter

Scatter plots are used to display 2 sets of variables on the 1 axis. For demonstration purposes I'm going to create some random data to plot. A lot of the same parameters are available to us as above and I am going to incorporate a few new ones.

```python
# using default_rng() to get a new random number generator
rng = np.random.default_rng()

# creates 2 variables with arrays of 100 random numbers between 1 and 100
x = rng.normal(1, 100, size=100)
y = rng.normal(1, 100, size=100)

# creates a basic scatter plot
plt.scatter(x,y)
```

To get something a bit more interesting we need another set of variables to plot. I've created another set of random data - this time using the uniform distribution and plotted them on the same axis.

```python
a = rng.uniform(1, 100, size=50)
b = rng.uniform(1, 100, size=50)

plt.scatter(x,y)
plt.scatter(a,b)
```

As I did in the plot() example I'm going to add a load of other manipulations just to demonstrate some of the possibilities. I'm not expecting marks for artistic interpretation here!

```python
# set the figure size
plt.figure(figsize=(12, 8))

# sets the x and y labels, specifies their sizes, font and colour
plt.xlabel("x axis", size=20, fontname="Herculanum", fontsize=20, color='navy')
plt.ylabel("y axis", size=20, fontname="Herculanum", fontsize=20, color='navy')

# sets the title and specifies size, font and colour
plt.title("Scatter Plot", size=20, fontname="Chalkduster", color='firebrick')

# creates a background grid and specifies the colour and transparency
plt.grid(True, color='darkgrey', alpha=0.5)

# creates the scatter plot. specifies colour, marker size (s), transparency (alpha), 
# marker style (marker), and edgecolor
plt.scatter(x,y, color='lightsteelblue', s=120, alpha=0.8, marker='*', edgecolors='k')
plt.scatter(a,b, color='lightcoral', s=80, alpha=0.6, marker='o', edgecolors='k')
```

We can also use the marker size and colour to respresent other variables thus creating multi-dimensional plots. In this plot, adapted from W3 Schools (https://www.w3schools.com/python/matplotlib_scatter.asp), as well as the values on the x and y axes, we have the marker colour and size representing 2 other variables (again created randomly here). I have tried to incorporate some other elements here including a colormap, some font editing, and some tick editing.

```python
# set the figure size
plt.figure(figsize=(15, 10))

# sets the x and y labels, specifies their sizes, font and colour
plt.xlabel("x axis", size=20, fontname="Comic Sans MS", fontsize=20, color='teal', loc='left')
plt.ylabel("y axis", size=20, fontname="Comic Sans MS", fontsize=20, color='teal', loc='top')

# sets the title and specifies size, font and colour
plt.title("Scatter Plot", size=20, fontname="Comic Sans MS", color='teal', fontweight='bold', loc='right')

colors = rng.uniform(1, 100, size=100)
sizes = 12 * (rng.uniform(1, 100, size=100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')

plt.colorbar()

plt.xticks(rotation=45)
plt.yticks(rotation=45)

plt.grid(alpha=0.5, linestyle='--')

plt.show()
```

```python
# experimenting with formatting
t = np.sin(np.linspace(-3,3,50))
# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', color="red")
plt.plot(t, t**2, 'bs', color="blue")
plt.plot(t, t**3, 'g^', color="green")
plt.plot(t, t**4, "o", color="orange")
plt.plot(t, t**5, "o-", color="black")
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.2)

plt.figure()

plt.subplot(2,2,1)
plt.plot(t1, f(t1), 'black')

plt.subplot(2,2,2)
plt.plot(t2, np.tan(2*np.pi*t2), 'r--')

plt.subplot(2,2,3)
plt.plot(t2, np.exp(t2), 'g^')

plt.subplot(2,2,4)
plt.plot(t2, np.cos(2*np.pi*t2), 'orange');
plt.show()
```

#### Ex. 9

```python
# adapted from: https://matplotlib.org/stable/tutorials/introductory/pyplot.html#sphx-glr-tutorials-introductory-pyplot-py

# make up some data in the open interval (0, 1)
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure()

# linear
plt.subplot(221)
plt.plot(x, y, color="red")
plt.yscale('linear')
plt.title('linear',color="red")
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y, color="green")
plt.yscale('log')
plt.title('log',color="green")
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean(),color="blue")
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog',color="blue")
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y, color="orange")
plt.yscale('logit')
plt.title('logit',color="orange")
plt.grid(True)
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)
plt.tight_layout()
plt.show()
```
